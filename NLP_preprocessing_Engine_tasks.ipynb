TASK 1 Conceptual Understanding

1. What is the difference between "Love" and "love" in NLP?

In Natural Language Processing, "Love" and "love" can be treated as two different tokens in a case-sensitive system. This means the model may interpret them as separate words, which can lead to inconsistency in analysis. To avoid this issue, a common preprocessing step is converting all text to lowercase, ensuring that words like "Love" and "love" are treated uniformly and improving model accuracy.

2. What happens if stopwords are not removed?

If stopwords (commonly used words like "is", "the", "and") are not removed, the dataset may contain unnecessary noise. This increases the dimensionality of the data and can make the model less efficient. It may also reduce performance because these words usually do not carry significant meaning and can distract the model from learning important patterns.

3. Provide two real-world scenarios where removing stopwords can be harmful.

Sentiment Analysis:
In phrases like "not good" or "not happy", removing the stopword "not" completely changes the meaning of the sentence, leading to incorrect sentiment classification.
Question Answering Systems:
Words like "what", "who", "where", and "how" are essential for understanding the intent of a question. Removing these stopwords can make the sentence meaningless and affect the system's ability to generate correct answers.

4. Explain the difference between stemming and lemmatization.

Stemming is a simple and fast process that reduces words to their root form by cutting off prefixes or suffixes. However, it may produce incorrect or non-dictionary words (e.g., "running" → "runn").
Lemmatization is a more advanced and accurate process that converts words into their base or dictionary form (lemma) using vocabulary and grammar rules (e.g., "running" → "run"). It ensures meaningful and valid words but requires more computational resources.

TASK 2 Build Advanced Preprocessing Function


```python
import re

def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # Remove email-like patterns
    text = re.sub(r'\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Handle repeated characters (soooo → so, goooood → good)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Remove special characters (keep alphabets and spaces)
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Remove very short tokens (≤2), except "no" and "not"
    tokens = text.split()
    filtered_tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]

    return ' '.join(filtered_tokens)


# ---- MAIN EXECUTION ----
if __name__ == "__main__":
    text = "WOW!!! This is soooo goooood!!! Visit http://example.com now, I have 2 dogs!!! Contact me at test@gmail.com"

    result = preprocess_text(text)

    print("Original Text:")
    print(text)

    print("\nProcessed Text:")
    print(result)
```

TASK 3 STRESS TESTING

```python
import re

def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # Remove email-like patterns
    text = re.sub(r'\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Handle repeated characters (soooo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Remove special characters (including emojis)
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Token filtering (≤2 removed except "no", "not")
    tokens = text.split()
    filtered_tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]

    return filtered_tokens, ' '.join(filtered_tokens)


# ---- TEST CASES ----
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

# ---- RUN TESTS ----
for i, sentence in enumerate(test_sentences, 1):
    tokens, cleaned = preprocess_text(sentence)

    print(f"\n--- Test Case {i} ---")
    print("Original Text   :", sentence)
    print("Cleaned Tokens  :", tokens)
    print("Cleaned Sentence:", cleaned)
```

TASK 4 TOKEN ANALYTICS


```python
import numpy as np

for i, sentence in enumerate(test_sentences):
    tokens, cleaned_sentence = preprocess_text(sentence)

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))

    avg_length = (sum(len(word) for word in tokens) / total_tokens) if total_tokens > 0 else 0

    print("Original Sentence:", sentence)
    print("Cleaned Sentence:", cleaned_sentence)
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Average Token Length:", round(avg_length, 2))
    print("-" * 50)
```

TASK 5 FREQUENCY ANALYSIS

```python
from collections import Counter

all_tokens = []

for sentence in test_sentences:
    tokens, cleaned_sentence = preprocess_text(sentence)
    all_tokens.extend(tokens)

word_freq = Counter(all_tokens)

print("Top 10 Most Frequent Words:")
print(word_freq.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(word_freq.most_common()[-5:])
```

TASK 6 BUILD FULL PIPELINE


```python

def full_pipeline(text_list):

    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, cleaned_sentence = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(cleaned_sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }



test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]


result = full_pipeline(test_sentences)

print("\n===== FINAL OUTPUT =====\n")
print(result)


```
